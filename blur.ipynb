```python
from CoRe_Dataloader_From_File_With_Random_From_Tensors import get_new_test_train_validation_datasets  
import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
_,_,dl = get_new_test_train_validation_datasets(0.1,0.05)
gb = transforms.GaussianBlur(41, sigma=(10, 10))
relu = torch.nn.ReLU()

def power(t):
    return ((torch.mean(t**2)) ** 0.5).item()
```

```python
data,params = dl[243]
```

```python
data.shape
```

```python
timg = data

plt.title("Original Image")
plt.imshow(timg.cpu().numpy())
plt.colorbar()
plt.show()
blurimg = torch.squeeze(gb(torch.unsqueeze(timg,0)))
print(power(timg),power(blurimg))
diff = power(timg)/power(blurimg)
print(diff)
blurimg *= diff
plt.imshow(blurimg.cpu().numpy())
plt.title("Blurred Image")
plt.colorbar()
plt.show()
plt.title("Difference")
sub = relu(timg-blurimg)
plt.imshow(sub.cpu().numpy())
plt.colorbar()
plt.show()
```
